```python
# Посчитаем уточним "экспертную оценку" доли аномалий в данных
anomaly_rate = df["Class"].sum()/df.shape[0]
anomaly_rate
```

# начинаем поиск аномалий

### 3 сигмы

```python
n_sigma = 18
tmp = df.drop(columns = ["Time", 'Class']).copy()
lower_bound = tmp.mean() - n_sigma * tmp.std()
upper_bound = tmp.mean() + n_sigma * tmp.std()
anomaly_3sigma = ((tmp<lower_bound) | (tmp>upper_bound)).any(axis=1)
anomaly_3sigma.sum()/anomaly_3sigma.shape[0]
```

### 1.5 IQR

```python
n_iqr = 43
iqr = tmp.quantile(0.75) - tmp.quantile(0.25)
lower_bound = tmp.mean() - n_iqr * iqr
upper_bound = tmp.mean() + n_iqr * iqr
anomaly_IQR = ((tmp<lower_bound) | (tmp>upper_bound)).any(axis=1)
anomaly_IQR.sum()/anomaly_IQR.shape[0]
```

### Distance-based

```python
from sklearn.base import BaseEstimator
from scipy.spatial.distance import cdist
from sklearn.preprocessing import RobustScaler

class DistanceOutliers(BaseEstimator):
    """
    Distance based outlier detector model
    
    Fit method calculates centroid of training samples and
    using `metric` get distances from centroid to train samples.
    Having distances - we calculate `threshold` based on `percentile`.
    
    Predict method uses `threshold` and `metric` to determine, whether
    distance to sample from centroid is too large -> outlier.
    
    -----------
    Parameters:
    
    - metric: string, default - euclidean
        metric to use for distance calculation (see scipy.spatial.distance.cdist)
        
    - percentile: float in range [0, 100]
        hyperparameter which sets the threshold for anomalies
    """
    def __init__(self, metric='euclidean', percentile=90):
        self.metric = metric
        self.percentile = percentile
        
    def fit(self, X):
        self.centroid = np.mean(X, axis=0).values.reshape(-1, 1).T
        distances_train = cdist(self.centroid, X, metric=self.metric).reshape(-1)
        self.threshold = np.percentile(distances_train, self.percentile)
        
    def predict(self, X):
        distances = cdist(self.centroid, X, metric=self.metric).reshape(-1)
        predictions = (distances > self.threshold).astype(int)
        return predictions
```

```python
scaler = RobustScaler()
scaled_data = pd.DataFrame(data=scaler.fit_transform(tmp), columns=tmp.columns)
```

#### евклидово расстояние

```python
euclidian_model = DistanceOutliers(metric='euclidean', percentile=99.84)
euclidian_model.fit(scaled_data)
euclidian_outliers = euclidian_model.predict(scaled_data)
euclidian_outliers.sum()/euclidian_outliers.shape[0]
```

#### манхэтановское расстояние

```python
citiblock_model = DistanceOutliers(metric='cityblock', percentile=99.83)
citiblock_model.fit(scaled_data)
cityblock_outliers = citiblock_model.predict(scaled_data)
cityblock_outliers.sum()/cityblock_outliers.shape[0]
```

### One-Class SVM

```python
%%time
from sklearn.svm import OneClassSVM
one_class_svm = OneClassSVM(nu=anomaly_rate, gamma='auto')
one_class_svm.fit(scaled_data.sample(50000))
svm_outliers = one_class_svm.predict(scaled_data)
svm_outliers = np.array([1 if label == -1 else 0 for label in svm_outliers])
svm_outliers.sum()/svm_outliers.shape[0]
```

```python
### IsolationForest
```

```python
%%time
from sklearn.ensemble import IsolationForest

isolation_forest = IsolationForest(n_estimators=100, contamination=anomaly_rate, 
                                   max_features=1.0, bootstrap=True, behaviour="new")
isolation_forest.fit(scaled_data.sample(50000))

isolation_outliers = isolation_forest.predict(scaled_data)
isolation_outliers = np.array([1 if label == -1 else 0 for label in isolation_outliers])
isolation_outliers.sum()/isolation_outliers.shape[0]
```

### Попробуем парочку интересных классов из библиотеки pyod

```python
%%time
from pyod.models.hbos import HBOS
from pyod.models.feature_bagging import FeatureBagging

hbos = HBOS(contamination=anomaly_rate)
fb = FeatureBagging(contamination=anomaly_rate, check_estimator=False,
                                          random_state=42)

hbos.fit(scaled_data.sample(50000))
fb.fit(scaled_data.sample(50000))

hbos_outliers = hbos.predict(scaled_data)
fb_outliers =  fb.predict(scaled_data)
```

### Посмотрим, что у нас с качеством

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
anomaly_ds = [
anomaly_3sigma, anomaly_IQR, euclidian_outliers, cityblock_outliers, 
svm_outliers, isolation_outliers, hbos_outliers, fb_outliers
]
anomaly_ds_name = [
"anomaly_3sigma", "anomaly_IQR", "euclidian_outliers", "cityblock_outliers", 
"svm_outliers", "isolation_outliers", "hbos_outliers", "fb_outliers"
]
```

```python
for i in range(len(anomaly_ds_name)):
    print("Report for {}\n\n".format(anomaly_ds_name[i])) 
    print(classification_report(df["Class"], anomaly_ds[i]))
    print(confusion_matrix(df["Class"], anomaly_ds[i]))
    print("\n")
```
